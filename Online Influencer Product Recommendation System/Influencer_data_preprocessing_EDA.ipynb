```python
# Since the views data are inaccurate we decided to set it aside for now and handle outliers in likes and comments
print(f'Post dataframe shape Before outlier handling: {df_posts.shape}')

def remove_outliers(df, col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Apply outlier removal for likes and comments
for col in ['likes', 'comments']:
  df_posts = remove_outliers(df_posts, col)

print(f'Post dataframe shape After outlier handling: {df_posts.shape}')
```

```python
# Feature Engineering: Extract total likes, views, comments and average likes, views, comments for each user_id
df_engagement = df_posts[['user_id', 'likes', 'comments', 'views']].copy()

# Calculate total and average engagement metrics per user_id
df_engagement_aggregated = df_engagement.groupby('user_id').agg(
    total_likes = ('likes', 'sum'),
    total_comments = ('comments', 'sum'),
    total_views = ('views', 'sum'),
    avg_likes_per_post = ('likes', 'mean'),
    avg_comments_per_post = ('comments', 'mean'),
    avg_views_per_post = ('views', 'mean')
).reset_index()

df_engagement_aggregated.head(5)
```

```python
import re
import nltk
import time
from tqdm import tqdm
from langdetect import detect
from nltk.corpus import stopwords
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from deep_translator import GoogleTranslator

nltk.download("stopwords")
nltk.download("punkt")
```

```python
def detect_language(text):
    """
    Detects the language of a given text.
    Returns the language code (e.g., 'en' for English).
    If detection fails, returns 'unknown'.
    """
    try:
        return detect(text)
    except:
        return "unknown"
```

```python
def translate_to_english(text):
    """
    Translates text to English if it's not already in English.
    Uses GoogleTranslator for automatic language detection and translation.
    """
    try:
        detected_lang = detect_language(text)
        if detected_lang != "en" and detected_lang != "unknown":
            return GoogleTranslator(source="auto", target="en").translate(text)
        return text  # Return original text if already in English
    except:
        return text  # In case of translation failure, return original text
```

```python
# Step 1: Detect language of captions
df_posts_encoded["language"] = df_posts_encoded["caption"].apply(detect_language)

# Step 2: Translate non-English captions to English
df_posts_encoded["translated_caption"] = df_posts_encoded["caption"].apply(translate_to_english)

# Step 3: Re-detect language after translation
df_posts_encoded["language"] = df_posts_encoded["translated_caption"].apply(detect_language)

# Step 4: Drop rows that are still non-English after translation
df_posts_encoded = df_posts_encoded[df_posts_encoded["language"] == "en"]

# Display the number of remaining English captions
print(f"Number of English captions retained: {df_posts_encoded.shape[0]}")
print(df_posts_encoded["language"].value_counts())
```

```python
def clean_text(text):
    """
    Cleans text by:
    - Removing HTML tags
    - Removing punctuation and special characters
    - Converting to lowercase
    - Removing stopwords
    """
    if not isinstance(text, str):
        return ""

    text = re.sub(r"<.*?>", "", text)  # Remove HTML tags
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove special characters and punctuation
    text = text.lower().strip()  # Convert to lowercase and strip spaces

    # Remove stopwords
    text = " ".join([word for word in text.split() if word not in stopwords.words("english")])

    return text
```

```python
df_posts_encoded["clean_caption"] = df_posts_encoded["translated_caption"].apply(clean_text)
```

```python
# You can continue with your NLP tasks such as:
# - Sentiment analysis using TextBlob or VADER
# - Keyword extraction
# - Topic modeling
# Example: Applying VADER sentiment analysis

analyzer = SentimentIntensityAnalyzer()

df_posts_encoded["sentiment_score"] = df_posts_encoded["clean_caption"].apply(lambda text: analyzer.polarity_scores(text)["compound"])

# Display a sample of results
df_posts_encoded[["clean_caption", "sentiment_score"]].head(10)
```
