```python
import selenium
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
#Initialize WebDriver
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service,options=chrome_options)
```

```python
# Selection criteria:
## Account type = Creator, Regular
## Content type = Has sponsored posts
# click"Show influencers"
```

```python
#Vist the Not Common website
# 1.scrape the first page
url = "https://notcommon.com/most-followed/instagram"
driver.get(url)
time.sleep(3)
name_list = [i.text for i in driver.find_elements(By.XPATH, ".//ul[@role='list']/li/div[position()=1]/div[position()=1]/a/span[@itemprop='name']")]
handle_list = [i.text for i in driver.find_elements(By.XPATH, ".//ul[@role='list']/li/div[position()=1]/div[position()=2]/a")]
follower_num_list = [i.text for i in driver.find_elements(By.XPATH, ".//ul[@role='list']/li/div[position()=2]/span/span[@class='tracking-wider']")]

#2.scape the following pages
for page in range(2,201):
    # open new page
    new_url = "https://notcommon.com/most-followed/instagram?page={}".format(page)
    driver.get(new_url)
    time.sleep(3)
    # extract data
    new_name_list = [i.text for i in driver.find_elements(By.XPATH, ".//ul[@role='list']/li/div[position()=1]/div[position()=1]/a/span[@itemprop='name']")]
    new_handle_list = [i.text for i in driver.find_elements(By.XPATH, ".//ul[@role='list']/li/div[position()=1]/div[position()=2]/a")]
    new_follower_num_list = [i.text for i in driver.find_elements(By.XPATH, ".//ul[@role='list']/li/div[position()=2]/span/span[@class='tracking-wider']")]
    # store new data to list
    name_list.extend(new_name_list)
    handle_list.extend(new_handle_list)
    follower_num_list.extend(new_follower_num_list)
    
    print(page)
```
