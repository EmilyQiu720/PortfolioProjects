```python
# Use TF-IDF for analyzing the caption column
df = df.dropna(subset = ['caption'])
df['caption'] = df['caption'].astype(str)
```

```python
from langdetect import detect, DetectorFactory
from collections import Counter
from langdetect.lang_detect_exception import LangDetectException
```

```python
# Retain only posts that's in english or spanish for simplicity
DetectorFactory.seed = 0

# Function to detect language with error handling
def safe_detect(text):
    try:
        return detect(text) if isinstance(text, str) and len(text.strip()) > 3 else 'unknown'
    except LangDetectException:
        return 'unknown'

# Apply language detection
df['language'] = df['caption'].apply(safe_detect)

# Count occurrences of each language
language_counts = Counter(df['language'])

# Convert to DataFrame for display
language_df = pd.DataFrame(language_counts.items(), columns=['Language', 'Count'])
```

```python
df = df[df['language'].isin(['en', 'es'])]
```

```python
from stopwordsiso import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
custom_stopwords = stopwords("en") | stopwords("es")  

vectorizer = TfidfVectorizer(stop_words=list(custom_stopwords), max_features=5000)
tfidf_matrix = vectorizer.fit_transform(df['caption'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=df['post_id'])
```

```python
cosine_sim_matrix = cosine_similarity(tfidf_matrix)
```

```python
cosine_sim_df = pd.DataFrame(cosine_sim_matrix, index=df['post_id'], columns=df['post_id'])
cosine_sim_df
```

```python
def get_top_similar_posts(post_id, top_n=5):
    similar_posts = cosine_sim_df[post_id].sort_values(ascending=False).iloc[1:top_n+1]
    return similar_posts
example_post_id = df['post_id'].iloc[18]  # Use the first post
top_similar_posts = get_top_similar_posts(example_post_id, top_n=10)
top_similar_posts
```

```python
# Normalize follower counts to a 0-1 range
user_follower_counts = df.groupby('user_id')['follower_count'].mean()

min_followers, max_followers = user_follower_counts.min(), user_follower_counts.max()
normalized_followers = (user_follower_counts - min_followers) / (max_followers - min_followers)

# Convert to similarity matrix
user_ids = normalized_followers.index
follower_similarity_matrix = 1 - np.abs(
    normalized_followers.values[:, None] - normalized_followers.values[None, :]
)
```

```python
follower_similarity_df = pd.DataFrame(follower_similarity_matrix, index=user_ids, columns=user_ids)
```

```python
follower_similarity_df
```

```python
# Now we have user_similarity_matrix based on post and follower_similarity_matrix based on follower
# Blending them to get final similarity matrix
alpha = 0.7  # Weight for user similarity
beta = 0.3   # Weight for follower similarity
adjusted_user_similarity = (alpha * user_similarity_matrix) + (beta * follower_similarity_matrix)
adjusted_user_similarity_df = pd.DataFrame(adjusted_user_similarity, index=user_ids, columns=user_ids)
```

```python
adjusted_user_similarity_df
```

```python
# Use KNN to find k (k=10) most similar influencers for each influencer
from sklearn.neighbors import NearestNeighbors
user_vectors = adjusted_user_similarity_df.values
# Fit KNN model
k_neighbors = 11  # Define the number of similar users to find
knn = NearestNeighbors(n_neighbors=k_neighbors, metric='cosine', algorithm='auto')
knn.fit(user_vectors)
user_id_map = {i: user_id for i, user_id in enumerate(adjusted_user_similarity_df.index)}
# Find nearest users for each user
distances, indices = knn.kneighbors(user_vectors)

# Create a dictionary of most similar users, excluding self (first entry)
similar_users_dict = {
    user_id_map[i]: [user_id_map[idx] for idx in indices[i][1:k_neighbors]]  # Exclude first (self)
    for i in range(len(user_vectors))
}
# Find nearest users for each user
distances, indices = knn.kneighbors(user_vectors)

# Create a dictionary of most similar users, excluding self (first entry)
similar_users_dict = {
    user_id_map[i]: [user_id_map[idx] for idx in indices[i][1:k_neighbors]]  # Exclude first (self)
    for i in range(len(user_vectors))
}
# Convert to DataFrame
similar_users_df = pd.DataFrame.from_dict(similar_users_dict, orient='index', columns=[f"Top_{i+1}_User" for i in range(10)])
```

```python
similar_users_df
```

```python
similarity_scores_dict = {
    user_id_map[i]: [1 - distances[i][j] for j in range(1, k_neighbors)]  # Convert cosine distance to similarity
    for i in range(len(user_vectors))
}
similarity_scores_df = pd.DataFrame.from_dict(similarity_scores_dict, orient='index', columns=[f"Top_{i+1}_Sim" for i in range(10)])
similarity_scores_df
```

```python
# Step 1: Convert wide format to long format for similar users
similar_users_long = similar_users_df.reset_index().melt(id_vars=['index'], var_name="rank", value_name="similar_user_id")
similar_users_long.rename(columns={'index': 'user_id'}, inplace=True)

# Normalize rank column by removing "_User"
similar_users_long['rank'] = similar_users_long['rank'].str.replace('_User', '', regex=False)

# Step 2: Convert wide format to long format for similarity scores
similarity_scores_long = similarity_scores_df.reset_index().melt(id_vars=['index'], var_name="rank", value_name="similarity_score")
similarity_scores_long.rename(columns={'index': 'user_id'}, inplace=True)

# Normalize rank column by removing "_Sim"
similarity_scores_long['rank'] = similarity_scores_long['rank'].str.replace('_Sim', '', regex=False)

# Step 3: Merge similar users with similarity scores using the normalized 'rank'
similar_users_expanded = similar_users_long.merge(
    similarity_scores_long, on=['user_id', 'rank'], how='left'
)
```

```python
# Step 5: Get products promoted by similar users
similar_users_products = df.merge(
    similar_users_expanded, left_on='user_id', right_on='similar_user_id'
)
```

```python
similar_users_expanded
```

```python
# Step 1: Create a mapping of user_id to their advertised products (category, name, brand) as sets for fast lookup
df['product_tuple'] = list(zip(df['ad_product_category'], df['ad_product_name'], df['advertising_brand']))
user_product_map = df.groupby('user_id')['product_tuple'].apply(set).to_dict()

# Step 2: Compute engagement scores once using groupby for efficiency
engagement_scores = df.groupby(['user_id', 'ad_product_category', 'ad_product_name', 'advertising_brand'])['normalized_engagement_score'].sum().reset_index()

# Step 3: Create recommendations list
recommendations = []

# Iterate over each user and their similar users
for _, row in similar_users_expanded.iterrows():
    target_user = row['user_id']
    similar_user = row['similar_user_id']
    similarity_score = row['similarity_score']
    
    # Get product sets for target and similar users
    similar_user_products = user_product_map.get(similar_user, set())
    target_user_products = user_product_map.get(target_user, set())

    # Find new products promoted by similar users but not the target user
    new_products = similar_user_products - target_user_products

    # Filter engagement scores efficiently
    if new_products:
        filtered_scores = engagement_scores[
            (engagement_scores['user_id'] == similar_user) &
            (engagement_scores[['ad_product_category', 'ad_product_name', 'advertising_brand']].apply(tuple, axis=1).isin(new_products))
        ]

        # Append recommendations
        for _, product_row in filtered_scores.iterrows():
            recommendations.append({
                'user_id': target_user,
                'similar_user_id': similar_user,
                'ad_product_category': product_row['ad_product_category'],
                'ad_product_name': product_row['ad_product_name'],
                'advertising_brand': product_row['advertising_brand'],
                'engagement_score': product_row['normalized_engagement_score'],
                'similarity_score': similarity_score,
                'recommendation_score': similarity_score * product_row['normalized_engagement_score']
            })

# Convert recommendations to a DataFrame
recommendations_df = pd.DataFrame(recommendations)
```

```python
# Drop rows where all three columns are 'na'
filtered_df = recommendations_df[
    ~((recommendations_df['ad_product_category'] == 'na') &
      (recommendations_df['ad_product_name'] == 'na') &
      (recommendations_df['advertising_brand'] == 'na'))
]
```

```python
# Aggregate the result: each user_id, each product (same category, name, brand) as a row
aggregated_df = filtered_df.groupby(
    ['user_id', 'ad_product_category', 'ad_product_name', 'advertising_brand'],
    as_index=False
)['recommendation_score'].sum()
```
